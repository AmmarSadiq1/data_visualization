## customizing plots in python

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

sns.set_theme(style="ticks", color_codes=True)
kashti = sns.load_dataset("titanic")
# print(kashti.head())
sns.boxplot(x="survived", y="age", showmeans=True, meanprops={
            "marker": "*", "markersize": "20", "markeredgecolor": "green"}, data=kashti)

#show labels
plt.xlabel("How many survived", size=18)
plt.ylabel("Age of people who survived", size=15)
plt.title("Survival data", size = 20, weight="bold")
plt.show()
```

### violin plot

```python
import seaborn as sns
from seaborn import palettes
import matplotlib.pyplot as plt

sns.set_theme(style="ticks", palette="pastel")
tip = sns.load_dataset("tips")

#drawing a violin plot and split the violins for easier comparison
sns.violinplot(x="day", y="total_bill", hue="smoker", data=tip, split=True, inner="quart",
               linewidth=1, palette={"Yes":"b", "No":"0.85"})
sns.despine(left=True)
plt.show()
```
